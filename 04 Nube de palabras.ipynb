```python
#@title Creamos el wordcloud, esto es solo para el fondo y algunos parámetros
wc = WordCloud(background_color='white',
               mask=my_mask,
               collocations=False,
               width=600,
               height=300,
               contour_width=3, # especifica el ancho del contorno
               contour_color='black', # especifica el color del contorno
               stopwords=palabras_irrelevantes) # agregar stopwords
```

```python
#@title Graficar la nube de palabras
plt.figure(figsize=(50, 25)) # Tamaño de la figura
plt.imshow(wc, interpolation='bilinear')
plt.axis('off') # Quito los ejes
wc.to_file('wordcloud.png') # Guardo el wordcloud
plt.show()
```
